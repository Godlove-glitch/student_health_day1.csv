```python
class PatientSurvey:
    def __init__(self,age,gender,anxiety,depression,academic_year):
        self.age = int (age)
        self.gender = gender
        self.anxiety = int (anxiety)
        self.depression = int (depression)
        self.academic_year = academic_year
    def show_summary(self):
        print (f"{self.gender}, {self.age}, |Anxiety level : {self.anxiety}/5, | depression level : {self.depression}/5,|self.academic_year : {self.academic_year}")
```

```python

sample = PatientSurvey(age=19, gender="Female", anxiety=4, depression=3, academic_year="Year 2")

sample.show_summary()
```
